### Check prediction format 

```python
# Filter rows where prediction is NOT in ['A', 'B', 'C', 'D']
prediction_column_names = ['multiple_choice_prediction_easy', 'multiple_choice_prediction_medium', 'multiple_choice_prediction_hard', 
                           'multiple_choice_prediction_Phi_easy', 'multiple_choice_prediction_Phi_medium', 'multiple_choice_prediction_Phi_hard']

for column_name in prediction_column_names:
    filtered_rows = prediction_result[~prediction_result[column_name].isin(['A', 'B', 'C', 'D'])]
    print(column_name, filtered_rows.shape[0])
```

### Prediction Accuracy

```python
def calculate_multiple_choice_question_accuracy(df):
    # Calculate accuracy
    # llava
    accuracy_easy = (df["multiple_choice_solution"] == df["multiple_choice_prediction_easy"]).mean()
    accuracy_medium = (df["multiple_choice_solution"] == df["multiple_choice_prediction_medium"]).mean()
    accuracy_hard = (df["multiple_choice_solution"] == df["multiple_choice_prediction_hard"]).mean()

    # Phi
    accuracy_easy_Phi = (df["multiple_choice_solution"] == df["multiple_choice_prediction_Phi_easy"]).mean()
    accuracy_medium_Phi = (df["multiple_choice_solution"] == df["multiple_choice_prediction_Phi_medium"]).mean()
    accuracy_hard_Phi = (df["multiple_choice_solution"] == df["multiple_choice_prediction_Phi_hard"]).mean()

    print('Prediction Accuracy for Llava: ') 
    print(f"Easy: {accuracy_easy * 100:.2f}%") 
    print(f"Medium: {accuracy_medium * 100:.2f}%") 
    print(f"Hard: {accuracy_hard * 100:.2f}%") 

    print('\n')
    print('Prediction Accuracy for Phi: ')
    print(f"Easy: {accuracy_easy_Phi * 100:.2f}%")
    print(f"Medium: {accuracy_medium_Phi * 100:.2f}%")
    print(f"Hard: {accuracy_hard_Phi * 100:.2f}%")

    return accuracy_easy, accuracy_medium, accuracy_hard, accuracy_easy_Phi, accuracy_medium_Phi, accuracy_hard_Phi
```

```python
calculate_multiple_choice_question_accuracy(prediction_result)
```

### Analyze failure cases

```python
# failure cases for llava easy
failure_cases_easy_llava = prediction_result[prediction_result["multiple_choice_solution"] != prediction_result["multiple_choice_prediction_easy"]]
```

```python
import re

# helper function to extract the choice from the question
def extract_choice(question: str, choice: str) -> str:
    # Build a regex pattern dynamically to match "choice." or "choice)"
    pattern = rf"\b{choice}[.)]\s(.*)"
    match = re.search(pattern, question)
    # Return the matched choice text if found, otherwise return a default message
    return match.group(1) if match else "Choice not found."
```

```python
# For each failure case: 
# 1. open the image using imread() given img_url, 
# 2. display the solution
# 3. extract the choices from the question

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for index, row in failure_cases_easy_llava.iterrows():
    # Load the image
    img = mpimg.imread(row['img_url'])
    
    # Display the image
    plt.imshow(img)
    plt.axis('off')
    plt.show()

    # print(f"Question: {row['multiple_choice_question_easy']}")
    # print("\n")

    # print(f"Prediction: {row['multiple_choice_prediction_easy']}")
    # print("\n")

    # Display the ground-truth solution
    print(f"Solution: {row['reference_caption'][0]}")
    #print("\n")

    # Extract the choices from the question
    print(f"Prediction: {extract_choice(row['multiple_choice_question_easy'], row['multiple_choice_prediction_easy'])}")

    print("\n")
    print("-----------------------------------------------------")
    
```
