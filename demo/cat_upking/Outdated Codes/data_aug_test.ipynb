```python
# Load annotation with multiple choice question data file
import pandas as pd
pd.set_option('display.max_colwidth', None)
import json


# Data Directory: 
#food_images_directory = '/shared/data/food_data/food_images/'
food_annotation_file_path = '/shared/data/upking/upking_annotation_with_MCQ.json'


# Method 1: Using pandas.read_json directly
df = pd.read_json(food_annotation_file_path)

df.head(1)
```

```python
def data_augumentation(image, type = None):
    if type == "flip":
        img = np.flip(image, axis=1)
        return img
    elif type == "blur":
        return image
    else:
        return image
```

```python
img_url = df.head(1)['img_url'][0]
img_url
```

```python
original_image = io.imread(img_url)
plt.imshow(original_image)
```

```python
flipped_image = data_augumentation(original_image, type = "flip")
plt.imshow(flipped_image)
```
